### Confusion Matrix Task

##### 가정: 고암(Abnormal) 환자와 정상(Normal) 환자를 분류하는 의료 진단 시나리오.

- 양성(Positive): 고암 환자를 의미.
- 음성(Negative): 정상 환자를 의미.

#### 문제 1: 유방암 진단

>True Positive (TP): 모델이 환자가 고암을 올바르게 분류한 경우 양성(True)
False Positive (FP): 모델이 환자가 정상인데 고암이라고 하는 경우 (error I)
True Negative (TN): 모델이 환자가 정상을 올바르게 분류한 경우 (True)
False Negative (FN): 모델이 환자가 고암인데 정상으로 잘못 분류한 경우 (error II)

>정밀도 (Precision): 모델이 환자가 고암으로 예측한 환자 중에서 실제 고암인지 환자의 비율을 나타낸다 즉, 양성으로 예측한 결과 중 얼마나 많은 것이 진짜 양성인지 측정한다.
재현율 (Recall): 실제 고암 환자 중에서 모델이 고암으로 정확하게 예측한 피해자의 비율을 나타낸다 즉, 모델이 얼마나 많은 고암 환자를 "놓치지 않고" 감지했는지 측정한다.
>무엇을 높여주는것이 적합한가?
>
> 이는 재현율를 높여야 하며, Threshold(임계값)을 높여서 FN를 낮추고(▼정밀도), FP을 올려야 한다(▲재현율). 
> 즉, Positive(고암)을 예측할 확률이 높아져야 사람의 생명을 구할 확률이 높아지는 것이다.
> 따라서 정밀도를 높여주면서 재현율를 낮춰주어야 된다.

##### 가정: 정상 메일과 스팸 메일을 분류하는 스팸 메일 분류 시나리오.

- 양성(Positive): 스팸 메일을 의미.
- 음성(Negative): 정상 메일을 의미.

#### 문제 2: 스팸 메일 필터링

> True Positive (TP): 모델이 스팸 메일을 올바르게 분류한 경우 양성(True) 
False Positive (FP): 모델이 스팸 메일인데 정상으로 하는 경우 (error I)
True Negative (TN): 모델이 메일이 정상인지 올바르게 분류한 경우
False Negative (FN): 모델이 정상인데 스팸메일로 잘못 분류한 경우 (error II)

> 정밀도 (Precision): 모델이 정상 메일인지 예측한 메일 중에서 실제 정상인 메일의 비율을 나타낸다 즉, 양성으로 예측한 결과 중 얼마나 많은것이 진짜 양성인지 측정한다.
재현율 (Recall): 실제 스팸 메일 중에서 모델이 스팸 메일인지 정확하게 예측한 피해자의 비율을 나타낸다 즉, 모델이 얼마나 많은 스팸 메일을 "놓치지 않고" 감지했는지 측정한다. 

> 이는 정밀도를 높여야 하며, Threshold(임계값)을 높여서 FP를 낮추고(▲정밀도), FN을 높여야 한다(▼재현율).
즉, Negative(Alive)로 예측할 확률이 높아져야 생명을 구할 확률이 높아지고, Threshold value(임계값)를 높여주는 것이다.
위의 시나리오처럼 한 쪽의 수치를 강제로 높이면 다른 한 쪽의 수치는 떨어지며,
이를 정밀도와 재현율의 "트레이드 오프"라고 한다.
임계치가 높아지면 정밀도가 높아진다. 반대로 임계치가 낮아지면 재현율이 높아진다.

##### 가정: 범죄 예측 시나리오.

- 양성(Positive): 범죄가 예측되는 경우 (예: 범죄가 발생함).
- 음성(Negative): 범죄가 예측되지 않는 경우 (예: 범죄가 발생하지 않음).

#### 문제 3: 범죄 예측

> True Positive (TP): 모델이 범죄인지 올바르게 분류한 경우 
False Positive (FP): 모델이 범죄로 예측되는데 범죄로 예측되지 않는 경우 (error I)
True Negative (TN): 모델이 범죄로 예측되지 않는 경우
False Negative (FN): 모델이 범죄로 예측되지 않는데 범죄로 예측되는 경우 (error II) 

> 정밀도 (Precision): 모델이 범죄로 예측되는지 예측한 것 중에서 실제 범죄일 비율을 나타낸다 즉, 범죄로 예측한 결과 얼마나 많은것이 진짜 범죄인지 측정한다. 
재현율 (Recall): 실제 범죄 중에서 모델이 범죄로 예측되지 않은지 정확하게 예측한 피해자의 비율을 나타낸다 즉, 모델이 얼마나 많은 범죄로 예측되지 않는 비율을 "놓치지 않고" 감지했는지 측정한다. 

> 이는 재현율를 높여야 하며, Threshold(임계값)을 낮추어서 FN를 낮추고(▼정밀도), FP을 높여야 한다(▲재현율). 
> 즉, Positive(범죄가 예측되는 경우)를 예측할 확률이 높아져야 범죄자를 많이 예측하면 그만큼 시민을 보호할 수 있는 수단이 높아지는 것이다.
> 따라서 재현율이 높여주어서 실제 범죄인지 올바르게 분류한 것 중에서 모델이 범죄인지 올바르게 분류한 것을 정확하게 예측한 범죄자의 비율을 나타낸다. 즉, 모델이 얼마나 많은 범죄자를 "놓치지 않고" 감지했는지 측정한다


##### 가정: 검색 엔진에서 웹 페이지의 검색 결과를 분류하는 검색 품질 모델 서비스 시나리오.

- 양성(Positive): 검색 결과가 사용자의 의도에 부합하는 경우 (예: 정확한 검색 결과).
- 음성(Negative): 검색 결과가 사용자의 의도에 부합하지 않는 경우 (예: 부정확한 또는 불량한 검색 결과).

#### 문제 4: 검색 품질 모델

> True Positive (TP): 모델의 검색 결과가 사용자의 의도에 올바르게 부합하는 경우 
False Positive (FP): 모델의 검색 결과가 사용자의 의도에 부합해야되는데 부합하지 않는 경우 (error I) 
True Negative (TN): 모델의 검색 결과가 사용자의 의도에 부합되지 않는 경우
False Negative (FN): 모델의 검색 결과가 사용자의 의도에 부합되지 말아야 하는데 잘못해서 의도로 부합하는 경우 (error II) 

> 정밀도 (Precision): 모델이 검색 결과가 사용자의 의도에 부합하는 것을 예측한 것 중에서 실제로 검색 결과가 의도적으로 부합하는 비율을 나타낸다. 즉 검색 결과가 얼마나 많이 의도적으로 부합됬는지 측정한다. 
재현율 (Recall): 실제로 정확한 검색 결과 중에서 모딜에 검색결과가 사용자의 의도에 부합하지 않는것인지 정확하게 예측한 비율을 나타낸다 즉, 검색 결과가 사용자의 의도에 얼마나 예측되지 않는 비율을 "놓치지 않고" 감지했는지 측정한다.

> 이는 정밀도를 높여야 하며, Threshold(임계값)을 높여서 FN를 낮추고(▼재현율), FP을 높여야 한다(▲정밀도). 
> 즉, Positive(모델의 검색 결과가 사용자의 의도에 올바르게 부합)를 예측할 확률이 높아지고 검색결과가 의도에 올바르게 부합되는것을 많이 예측하면 그만큼 검색 품질이 많이 향상되기 때문이다.
> 따라서 정밀도를 높여주면서 실제 검색 결과 중에서 모델이 검색결과로 사용자의 의도에 올바르게 부합되는지 정확하게 예측한 비율을 나타낸다. 즉 모델이 얼마나 정상메일인지 예측한 결과 중 얼마나 많은 것이 진짜 양성인지 측정하기 떄문이다.

##### 가정: 보안 시스템에서 침입 감지 시스템 시나리오.

- 양성(Positive): 침입이 감지된 경우 (예: 보안 이벤트가 발생함).
- 음성(Negative): 침입이 감지되지 않은 경우 (예: 정상 상황).  

#### 문제 5: 침입 감지 시스템

> True Positive (TP): 모델이 침입이 감지 되는 경우를 올바르게 표현한 경우 
False Positive (FP): 모델이 침입이 감지되어야 하는데 감지되지 않는 경우 (error I)
True Negative (TN): 모델이 침입이 감지 되지 않는 경우
False Negative (FN): 모델이 침입이 감지되지 말아야 하는데 오작동으로 감지가 되는 경우 (error II)

> 정밀도 (Precision): 모델이 침입이 감지 되어야 하는 것을 예측한 것중에서 실제로 침입이 감지되는 비율을 나타낸다. 즉 침입이 얼마나 감지되는지 측정한다.
재현율 (Recall): 실제 침입 중에서 모델이 감지 되지 않은 것을 예측한 것 중에서 정확하게 침입이 감지되지 않는 것의 비율을 나타낸다. 즉, 칩입이 얼마나 감지 되지 않는지를 예측되는지의 비율을 "놓치지 않고" 예측했는지 측정한다.

> 이는 정밀도를 높여야 하며, Threshold(임계값)을 높여서 FN를 낮추고(▼재현율), FP을 높어야 한다(▲정밀도). 
> 즉, Negative(침입이 감지 되지 않느 경우)을 예측할 확률이 높아져야 사람의 안전성이 보호가 더 되는 것이다.
> 따라서 재현율이 높여주면서 정밀도를 낮춰주어야 된다.

False Negative (FN): 모델이 정확한 검색 결과를 부정확한 검색 결과로 잘못 분류한 경우(잘못된 부정 예측, error II).  

> 정밀도 (Precision): 모델이 정확한 검색 결과로 분류한 검색 결과 중에서 실제로 정확한 검색 결과인 경우의 비율을 나타낸다. 즉, 양성 클래스로 예측한 결과 중 얼마나 많은 것이 진짜 양성(정확한 검색 결과)인지 측정한다.  
재현율 (Recall): 실제로 정확한 검색 결과 중에서 모델이 정확한 검색 결과로 정확하게 예측한 검색 비율을 나타낸다. 즉, 모델이 얼마나 많은 정확한 검색 결과를 "놓치지 않고" 감지했는지 측정한다.  

> 🏆임계치를 높여서 정밀도를 높여주는 것이 적합하다.

##### 가정: 보안 시스템에서 침입 감지 시스템 시나리오.

- 양성(Positive): 침입이 감지된 경우 (예: 보안 이벤트가 발생함).
- 음성(Negative): 침입이 감지되지 않은 경우 (예: 정상 상황).  

#### 문제 5: 침입 감지 시스템

> True Positive (TP): 모델이 침입을 정확하게 감지한 경우.  
False Positive (FP): 모델이 정상 상황을 침입으로 잘못 감지한 경우(오작동, error I).  
True Negative (TN): 모델이 정상 상황을 정상으로 정확하게 분류한 경우.  
False Negative (FN): 모델이 실제로 침입이 있음에도 불구하고 침입을 감지하지 못한 경우(놓침 error II).  

> 정밀도 (Precision): 모델이 침입으로 감지한 상황 중에서 실제로 침입이 발생한 상황의 비율을 나타낸다. 즉, 양성 클래스로 예측한 결과 중 얼마나 많은 것이 진짜 양성(침입)인지 측정한다.  
재현율 (Recall): 실제 침입 중에서 모델이 침입으로 정확하게 예측한 상황의 비율을 나타낸다. 즉, 모델이 얼마나 많은 침입을 "놓치지 않고" 감지했는지 측정한다.

> 🏆임계치를 낮추어서 재현율을 높여주는 것이 적합하다.
